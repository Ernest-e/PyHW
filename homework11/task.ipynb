```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x')
expr = sp.sin(x)**2 - sp.cos(x)**2
expr
```

```python
# Корни
expr_solve = sp.solve(expr) 
expr_solve
```

```python
expr_diff = sp.diff(expr)
expr_diff
```

```python
# возрастает
rise_exp = sp.solve(expr_diff > 0)
rise_exp
```

```python
# убывает
down_exp =  sp.solve(expr_diff < 0)
down_exp
```

```python
# вершины
peak_exp = sp.solve(expr_diff)
peak_exp
```

```python
# больше нуля
more_null = sp.solve(expr > 0)
more_null
```

```python
# меньше нуля
less_null = sp.solve(expr < 0)
less_null
```
